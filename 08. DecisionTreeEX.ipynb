# Decision Tree (결정트리)

#### 일련의 분류 규칙을 통해 데이터를 분류, 회귀하는 지도학습 모델 중 하나


## 작동원리 

#### Root Node로 부터 시작하여 분류기준(조건)에 따라 전체 영역을 두 개로 구분하며, 맨 마지막 노드를 Terminal Node 혹은 Leaf Node라고 한다. 

---

## sklearn.tree.plot_tree

### plot_tree(decision_tree, max_depth=None, feature_names=None, class_names=None, filled=False, proportion=False, rounded=False, precision=3, fontsize=None)

#### - max_depth: if None, the tree is fully generated.
#### - feature_names: list of strings
#### - class_names: list of str or bool
#### - proportion: True로 지정하면, 값 대신 비율로 표현
#### - precision: number of digits of precision
#### - filled, rounded, fontsize: 노드의 모양과 폰트크기 설정

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data[:, 2:] # 꽃잎의 길이와 너비
y = iris.target

clf = DecisionTreeClassifier(max_depth=2)  # max_depth : 분리의 최대 깊이 (규제를 가함)
clf.fit(X, y)   

plt.figure(dpi=120)   
plot_tree(clf, feature_names=iris.feature_names[2:],
          class_names=iris.target_names,
          filled=True, rounded=True)
plt.show()
```
