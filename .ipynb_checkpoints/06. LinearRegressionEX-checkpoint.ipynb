## cost 함수 

```python
import numpy as np
import matplotlib.pyplot as plt 

X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([1, 2, 3, 4, 5], dtype=float)

def run(W):
    hypothesis = W * X
    cost = np.mean((hypothesis-y)**2)
    return cost

W_val = []
cost_val = []

for i in range(-30, 50):
    w = i * 0.1
    c = run(w)
    
    if i%10 == 0:
        print(f'w = {w:5.1f}, c = {c:5.1f}')
        W_val.append(w)
        cost_val.append(c)

plt.plot(W_val, cost_val)
plt.show()
```

## Linear Regression 예제 

#### 임의의 데이터 준비 

```python
import numpy as np
import matplotlib.pyplot as plt

# random data
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# plot
plt.scatter(X, y)
plt.axis([0, 2, 0, 15])
plt.show()
```

### Linear Regression 적용

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# random data
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# linear regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
print(f"기울기 = {lin_reg.coef_}, y절편 = {lin_reg.intercept_}")

# plot
plt.scatter(X, y)
new_y = lin_reg.coef_ * X + lin_reg.intercept_

plt.plot(X, new_y, "r-")
plt.axis([0, 2, 0, 15])
plt.show()
```

## Boston 집값 예측하기

### 데이터 로드 후 데이터 프레임으로 변환 

```python
from sklearn.datasets import load_boston  #boston data 로드 
import pandas as pd

boston = load_boston() # boston 변수에 load_boston 함수 저장 
 # boston.data: 독립변수 ndarray 배열, boston.feature_names: 독립변수 이름 리스트
dfX = pd.DataFrame(boston.data, columns=boston.feature_names) # dfX = 독립변수 행렬
# boston.target: 종속변수 ndarray 배열
dfy = pd.DataFrame(boston.target, columns=["PRICE"]) #dfy = 종속변수 벡터, PRICE = 종속 변수 이름  
df = pd.concat([dfX, dfy], axis=1) # 독립변수와 종속 변수의 데이터 프레임을 하나의 데이터 프레임으로 묶는 작업 , axis=1 : 옆으로 붙이는 작업 
df # 출력 
```

### 보스턴 주택 가격 예측 (선형회귀)

```python
#선형회귀 
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
from sklearn import metrics

boston = load_boston()
X = boston.data
y = boston.target
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3)

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

y_predict = lin_reg.predict(X_test)
score = metrics.r2_score(y_test, y_predict)
print(f"accuracy = {score}")
```

### 보스턴 주택 가격 예측 (Ridge 선형회귀)

```python
#Ridge 선형회귀
from sklearn.datasets import load_boston
from sklearn.linear_model import Ridge
from sklearn import model_selection
from sklearn import metrics

boston = load_boston()
X = boston.data
y = boston.target
X_train, X_test, y_train, y_test= model_selection.train_test_split(X, y, test_size=0.3)

lin_reg = Ridge(alpha=1.0)
lin_reg.fit(X_train, y_train)

y_predict = lin_reg.predict(X_test)
score = metrics.r2_score(y_test, y_predict)
print(f"accuracy = {score}")
```

### 보스턴 주택 가격 예측 (Lasso 선형회귀)

```python
# Lasso선형회귀 
from sklearn.datasets import load_boston
from sklearn.linear_model import Lasso
from sklearn import model_selection
from sklearn import metrics

boston = load_boston()
X = boston.data
y = boston.target
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3)

lin_reg = Lasso(alpha=1.0)
lin_reg.fit(X_train, y_train)

y_predict = lin_reg.predict(X_test)
score = metrics.r2_score(y_test, y_predict)
print(f"accuracy = {score}")
```
