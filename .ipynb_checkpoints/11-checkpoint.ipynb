```python
import tensorflow as tf

x_data = [1, 2, 3, 4, 5]
y_data = [1.5, 2.0, 2.5, 3.0, 3.5]

W = tf.Variable(tf.random.normal([1]), name='Weight') # 난수를 1로 생성 = 벡터 하나, 
B = tf.Variable(tf.random.normal([1]), name='Bias')   # 편향 

@tf.function            # 가설 함수 정의 
def Hyper(X):           # Hypothesis
    return W * X + B    

@tf.function            # 비용 함수 정의 
def Loss(H, Y):         # loss
    return tf.reduce_mean(tf.square(H - Y))   #square : 뺄셈에 대한 제곱, reduce_mean : square 값을 더해 평균한 값 

@tf.function            # 학습 함수 정의 
def Train(X, Y, lr=0.01):     #train lr = learning_rate
    with tf.GradientTape() as tape:          # with절로 tensorflow에 gradientTape 객체를 생성 
        _loss = Loss(Hyper(X), Y)            # loss function에 예측값과 실제값을 넣기 
    _w, _b = tape.gradient(_loss, [W, B])    # 미분계수 구하기 
    W.assign_sub(lr * _w)         # 텐서의 값 변경(갱신)을 위한 메소드, lr * 미분값(W)
    B.assign_sub(lr * _b)         # assign(=), assign_add(+=), assign_sub(-=), lr * 미분값(W)

for step in range(2000):
    Train(x_data, y_data, lr=0.01)
    _c = Loss(Hyper(x_data), y_data)    # cost 값 계산 

    if step % 20 == 0:        #20번에 한번씩 값 출력 
        print(f"{step}: {_c.numpy()} {W.numpy()} {B.numpy()}")
#         print('\nfinal W =', W.numpy(), 'b =', B.numpy())

# #test data
# test_data = [2, 4, 1, 5, 3]
# for data in test_data:
# y = data * W.numpy() + B.numpy()
# print("X =", data, "then Y =", y)
```
