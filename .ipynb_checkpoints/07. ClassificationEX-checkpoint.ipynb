## 정규분포 나이브베이즈 (GassianNB)

```python
from sklearn.preprocessing import LabelEncoder     #문자를 정수형 숫자로 변환하는 모듈 

weather = ['Sunny','Sunny','Overcast','Rainy','Rainy', 
           'Rainy','Overcast','Sunny','Sunny','Rainy', 
           'Sunny','Overcast','Overcast','Rainy']
play = ['No','No','Yes','Yes','Yes', 
        'No','Yes','No','Yes','Yes', 
        'Yes','Yes','Yes','No']

# 라벨 인코더 생성
le = LabelEncoder()

X = le.fit_transform(weather)  #weather 데이터를 피팅하고 라벨 숫자로 변환 
y = le.fit_transform(play) #play 데이터를 피팅하고 라벨 숫자로 변환 

print(X) # 0:Overcast, 1:Rainy, 2:Sunny  / 알파벳 순으로 
print(y) # 0:No, 1:Yes
```

```python
from sklearn.naive_bayes import GaussianNB  #정규뷴포 나이브베이즈

model = GaussianNB()    

#1차원 데이터를 2차원으로 변환하기 위해 reshape 작업을 함 
model.fit(X.reshape(-1,1), y.reshape(-1,1))   #reshape(-1,1) : 열의 값이 특정 정수로 지정 되어있을 때, 남은 배열의 길이와 남은 차원으로부터 추정해서 지정

# Predict
predicted = model.predict([[0]]) # 0:Overcast : 흐린날 

print("Predicted Value:", predicted)  # 0:No, 1:Yes
```

## Perceptron을 python으로 직접 구현 

```python
import numpy as np
 
class Perceptron:
    #eta : 학습률(learning_rate), n_iter : 최저값을 찾을 때 이동하는 횟수(반복횟수), random_state : 재현률(난수 생성이 같은 패턴으로 반복)
    def __init__(self, eta= 0.01, n_iter = 50, random_state=1):  
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state
        
    def predict(self, X):
        hyperthesis = np.dot(X, self.w_[1:]) + self.w_[0]
        return np.where(hyperthesis> 0.0, 1, -1)   # 주어진 조건보다 크면 1, 아니면 -1 반환 
 
    #학습 알고리즘 정의 
    def fit(self, X, y):
        regen = np.random.RandomState(self.random_state)
        self.w_ = regen.normal(loc = 0.0, scale=0.01, size= 1 + X.shape[1])  #normal : 난수 발생 
        self.errors_ = []    # 빈 리스트 생성 
 
        for _ in range(self.n_iter):    
            errors = 0                  # 손실함수 값 측정을 위해 errors = 0
            for features, target in zip(X, y):
                update = self.eta * (target - self.predict(features))   # 정답과 예측값의 차이를 학습률과 곱함 
                self.w_[1:] += update * features
                self.w_[1:] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self
 
    def net_input(self, X):
        """ 최종 입력 계산"""
        return np.dot(X, self.w_[1:]) + self.w_[0]
 
    def predict(self, X):
        return np.where(self.net_input(X) >= 0.0, 1, -1)
```

### 데이터 준비 

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np

iris = load_iris()
X = iris["data"][0:100, (0, 2)] # 처음 100개의 데이터 중에서 꽃받침 길이, 꽃잎 길이
y = iris["target"][0:100] # 처음 50개는 Iris-setosa, 다음 50개는 Iris-versicolor
y = np.where(y==0, -1, 1) # 만일 Iris-setosa이면 -1, 아니면 1로 변경

plt.scatter(X[0:50, 0], X[0:50, 1], color="r", marker="o", label="setosa")
plt.scatter(X[50:100, 0], X[50:100, 1], color="b", marker="^", label="versicolor")
plt.xlabel('sepal length(cm)')
plt.ylabel('petal length(cm)')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
X = iris["data"][0:100, (0,2)] # 처음 100개의 데이터 중에서 꽃받침 길이, 꽃잎 길이
y = iris["target"][0:100] # 처음 50개는 Iris-setosa, 다음 50개는 Iris-versicolor
y = np.where(y==0, -1, 1) # 만일 Iris-setosa이면 -1, 아니면 1로 변경

p = Perceptron(eta=0.1, n_iter=10)     #eta : 학습률(learning_rate), n_iter : 최저값을 찾을 때 이동하는 횟수(반복횟수)
p.fit(X, y)  # 학습 

plt.plot(range(1, len(p.errors_)+1), p.errors_, "bo--")  # 1부터 10까지 반복해서 에러률 확인 
plt.xlabel('Epochs')
plt.ylabel('Number of errors')
plt.show()
```

### 참고 : 결정영역 그리기

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap #결정 경계 그래프 함수 

def plot_decision_region(X, y, classifier):
    
    # 마커와 컬러 맵을 설정
    markers = ('o', '^', 'x')
    colors = ('red', 'green', 'blue')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    #결정 경계 그리기 
    x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
    y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

    Z = classifier.predict(np.array([xx.ravel(), yy.ravel()]).T)
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=cmap)
    plt.axis([x_min, x_max, y_min, y_max])

    # 샘플의 산점도 그리기 
    for idx, target in enumerate(np.unique(y)):
        plt.scatter(X[y==target, 0], X[y==target, 1],
                    alpha=0.8, c=colors[idx], marker=markers[idx],
                    label=np.where(target==-1, 'setosa', 'versicolor'))
        
# iris 데이터 로드 
iris = load_iris()
X = iris["data"][0:100, (0,2)] # 처음 100개의 데이터 중에서 꽃받침 길이, 꽃잎 길이
y = iris["target"][0:100] # 처음 50개는 Iris-setosa, 다음 50개는 Iris-versicolor
y = np.where(y==0, -1, 1) # 만일 Iris-setosa이면 -1, 아니면 1로 변경

plot_decision_region(X, y, classifier=p)
plt.xlabel('sepal length(cm)')
plt.ylabel('petal length(cm)')
plt.legend()
plt.show()
```

### sklearn.linear_model.Perceptron 사용하기

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris["data"][0:100, (0,2)] # 처음 100개의 데이터 중에서 꽃받침 길이, 꽃잎 길이
y = iris["target"][0:100] # 처음 50개는 Iris-setosa, 다음 50개는 Iris-versicolor
y = np.where(y==0, -1, 1) # 만일 Iris-setosa이면 -1, 아니면 1로 변경

# x 데이터와 y데이터를 test 데이터, train 데이터로 분리 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

ppn = Perceptron(eta0=0.1, max_iter=30, random_state=1) 
ppn.fit(X_train, y_train)
y_pred = ppn.predict(X_test)

print(f"accuracy: {accuracy_score(y_test, y_pred)}")
```
